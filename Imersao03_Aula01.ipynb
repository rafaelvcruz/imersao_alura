```python
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
```

```python
dados.head()
```

```python
##Desafio 01: Investigar por que a classe tratamento é tão desbalanceada.

dados['tratamento'].value_counts(normalize = True)

dados_filtrados = dados[dados['tratamento'] == 'com_droga']
```

```python
### Resposta: Através do sum, é possível observar que o tipo de tratamento "Com Controle", foi observado apenas em um tipo de droga.
dados.groupby(["tratamento", "droga"]).sum()
```

```python
####Desafio 02: Plotar as 5 últimas linhas da tabela.
dados.tail()
```

```python
####Desafio 03: Proporção das classes tratamento.
dados['tratamento'].value_counts(normalize = True)
```

```python
###Desafio 04: Quantos tipos de drogas foram investigados.
dados.groupby(["droga"]).count()
```

```python
####Desafio 06: Renomear as colunas tirando o hifen
dados.columns = dados.columns.str.replace('-', '')
dados.head()
```

```python
###Desafio 07: Deixar o grafico bonito {matplotlib}
import matplotlib.pyplot as plt
names = dados['tempo'].unique()
values = dados['tempo'].value_counts()
plt.figure(figsize=(9, 3))
plt.bar(names,values)
plt.suptitle("Gráfico Bonitão")
```
